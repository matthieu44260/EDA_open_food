```python
# Selection du produit le plus sucré

produit_a_remplacer = df_exemple.sort_values('sugars_100g',ascending = False).head(1)
produit_a_remplacer
```

```python
# liste de produits du même groupe mais moins sucré et mieux noté

groupe = produit_a_remplacer.iloc[0]['pnns_groups_2']
sucre = produit_a_remplacer.iloc[0]['sugars_100g']
note = produit_a_remplacer.iloc[0]['nutrition_grade_fr']
df.loc[(df['pnns_groups_2'] == groupe) & (df['sugars_100g'] < sucre) & (df['nutrition_grade_fr'] < note), :]
```
